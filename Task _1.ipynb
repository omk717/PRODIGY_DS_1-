```python
gender_counts = df['Region'].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(gender_counts.index, gender_counts.values, width=0.9)
plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Distribution of Region')

plt.xticks(gender_counts.index, rotation=45)
plt.tight_layout()
plt.show()
```

```python
np.random.seed(42)
ages = np.random.randint(1, 100, size=1000)  
genders = np.random.choice(['Male', 'Female'], size=1000)  

data = pd.DataFrame({'Age': ages, 'Gender': genders})

plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='Gender', data=data)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
